# A3 - Using Rasterio to reproject data (if necessary)

To use the web aplication we are developing as our final project (https://github.com/python-amazing/python-final-project), the user will be able to upload their own data. To ensure compatibility between the models from TorchGeo, all input images must use the same coordinate reference system (CRS), such as EPSG:4326 (WGS84).

The purposes of this notebook are:

1. Open a raster using `rasterio`
2. Check its CRS
3. Reproject it to the desired CRS (if necessary)
4. Save the result
5. Compare the input and the output

This can be reused in the final project as a preprocessing step.

```python
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.enums import Resampling
from rasterio.plot import show
from rasterio import Affine
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
def reproject_to_target_crs(input_path, output_path, target_crs="EPSG:4326"):
    with rasterio.open(input_path) as src:
        if src.crs == target_crs:
            return "No reprojection needed."

        transform, width, height = calculate_default_transform(
            src.crs, target_crs, src.width, src.height, *src.bounds
        )

        output_metadata = src.meta.copy()
        output_metadata.update({
            'crs': target_crs,
            'transform': transform,
            'width': width,
            'height': height
        })

        with rasterio.open(output_path, 'w', **output_metadata) as dst:
            for band_index in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, band_index),
                    destination=rasterio.band(dst, band_index),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=target_crs,
                    resampling=Resampling.nearest
                )
        return f"Reprojected raster saved to: {output_path}"
```

```python
input_file = "example_input.tif"
output_file = "example_output.tif"

reproject_to_target_crs(input_file, output_file)
```

```python
def summarize_raster(path):
    with rasterio.open(path) as src:
        image = src.read(1)
        crs = src.crs.to_string()
        width = src.width
        height = src.height
        pixel_size = (src.transform[0], -src.transform[4])
        bounds = src.bounds
    return image, crs, (width, height), pixel_size, bounds

input_path = "example_input.tif"
output_path = "example_output.tif"

img1, crs1, size1, px1, bounds1 = summarize_raster(input_path)
img2, crs2, size2, px2, bounds2 = summarize_raster(output_path)

fig, axes = plt.subplots(2, 2, figsize=(14, 10), gridspec_kw={'height_ratios': [3, 1]})
plt.subplots_adjust(hspace=0.3)

# Row 1: images
axes[0, 0].imshow(img1, cmap='gray')
axes[0, 0].set_title("Original Raster")
axes[0, 0].axis('off')

axes[0, 1].imshow(img2, cmap='gray')
axes[0, 1].set_title("Reprojected Raster")
axes[0, 1].axis('off')

# Row 2: text info
axes[1, 0].axis('off')
axes[1, 1].axis('off')


text1 = f"""CRS: {crs1}
Size: {size1[0]} x {size1[1]} px
Bounds:
{bounds1.left:.2f}, {bounds1.bottom:.2f}
{bounds1.right:.2f}, {bounds1.top:.2f}
"""

text2 = f"""CRS: {crs2}
Size: {size2[0]} x {size2[1]} px
Bounds: 
{bounds2.left:.6f}, {bounds2.bottom:.6f}
{bounds2.right:.6f}, {bounds2.top:.6f}
"""

axes[1, 0].text(0, 0.5, text1, fontsize=11, va='center', ha='left', family='monospace')
axes[1, 1].text(0, 0.5, text2, fontsize=11, va='center', ha='left', family='monospace')

plt.show()
```
